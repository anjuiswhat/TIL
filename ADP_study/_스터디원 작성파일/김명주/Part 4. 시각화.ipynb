Question 14) 다음으로는 ?lines를 실행했을 때 제공되는 예제 코드를 살펴보자.    
cars 데이터는 자동차의 속도(speed)와 제동거리(dist)로 이루어져 있다.    
먼저 첫 번째 코드를 실행하여 cars 데이터의 두 변수에 대한 산점도를 생성한 뒤,    
두 번째 코드를 이용해 cars 데이터에 LOWESS를 적용한 선을 덧그려보자.

```python
import pandas as pd
cars_df=pd.read_csv(r'C:\ISLR-python-master\Notebooks\Data\Cars.csv', index_col='Unnamed: 0')
cars_df.head()
```

```python
scatter=plt.scatter(cars_df['speed'], cars_df['dist'])
plt.title('Stopping Distance versus Speed')
```

```python
from scipy.interpolate import interp1d
import statsmodels.api as sm
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
! pip install plotly
```

```python
x=cars_df['speed'].values
y=cars_df['dist'].values

lowess = sm.nonparametric.lowess(x, y, frac=.3)
lowess_x = list(zip(*lowess))[0]
lowess_y = list(zip(*lowess))[1]
f = interp1d(lowess_x, lowess_y, bounds_error=False)
xnew = [i/3. for i in range(400)]
ynew = f(xnew)
plt.plot(cars_df['dist'], cars_df['speed'], 'o')
plt.plot(xnew, ynew, '-')
plt.show()
```

참고: https://pythonq.com/so/python/409531

```python
# Generate y_hat values using lowess, try a couple values for hyperparameters
y_hat1 =  sm.nonparametric.lowess(x, y) # note, default frac=2/3
y_hat2 = sm.nonparametric.lowess(x, y, frac=1/5)
```

```python
# Create a scatter plot
fig = px.scatter(cars_df, y, x, 
                 opacity=0.8, color_discrete_sequence=['black'])

# Add the prediction line
fig.add_traces(go.Scatter(x=y_hat1[:,0], y=y_hat1[:,1], name='LOWESS, frac=2/3', line=dict(color='red')))
fig.add_traces(go.Scatter(x=y_hat2[:,0], y=y_hat2[:,1], name='LOWESS, frac=1/5', line=dict(color='orange')))

# Change chart background color
fig.update_layout(dict(plot_bgcolor = 'white'))

# Update axes lines
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgrey', 
                 zeroline=True, zerolinewidth=1, zerolinecolor='lightgrey', 
                 showline=True, linewidth=1, linecolor='black')

fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgrey', 
                 zeroline=True, zerolinewidth=1, zerolinecolor='lightgrey', 
                 showline=True, linewidth=1, linecolor='black')

# Set figure title
fig.update_layout(title=dict(text="LOWESS test figure with Plotly", 
                             font=dict(color='black')))

# Update marker size
fig.update_traces(marker=dict(size=3))

fig.show()
```

Question 16) 정규분포에서 x값에 따른 확률 밀도를 계산해주는 dnorm함수를 사용해 평균이 0이고,    
표준편차가 1인 확률밀도함수를 생성하고, curve함수를 사용해 이것을 그래프로 나타내자.    
x축의 범위는 –3~+3로 지정하자.   

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.xlim(-3, 3)
plt.show()
```

또한 legend인자를 사용하여 각 변수가 어떻게 그래프에 표현되어져 있는지를 보여주는 범례도 함께 나타내자.   

```python
# summary by group

tips_sum_by_cyl_non = cars93_df[cars93_df['Origin'] == 'non-USA'].groupby('Cylinders').count()
tips_sum_by_cyl = cars93_df[cars93_df['Origin'] == 'USA'].groupby('Cylinders').count()

# Stacked Bar Chart

label = ['non-USA', 'USA']
N = len(cars93_df['Cylinders'].unique())
index = np.arange(N)
alpha = 0.5


p1 = plt.bar(index, tips_sum_by_cyl_non, color='b', alpha=alpha)
p2 = plt.bar(index, tips_sum_by_cyl, color='r', alpha=alpha,
             bottom=tips_sum_by_cyl_non) # stacked bar chart

plt.title('Stacked Bar Chart of Sum of Tips by Day & Sex', fontsize=20)
plt.ylabel('Sum of Tips', fontsize=18)
plt.xlabel('Day', fontsize=18)
plt.xticks(index, label, fontsize=15)
plt.legend((p1[0], p2[0]), ('Male', 'Female'), fontsize=15)
plt.show()
```
