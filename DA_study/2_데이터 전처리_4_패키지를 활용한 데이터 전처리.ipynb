#### [Problem]  adply
- iris 데이터에서 Petal.Length 변수가 1.5미만이면서 Species 변수 값이 ‘setosa’인 조건을 만족하는 경우 ‘1’을 그렇지 않은 경우 ‘0’을 부여한 칼럼을 생성하여, 원래의 iris 데이터와 함께 데이터프레임 형태로 출력하라.

```python
import numpy as np
import pandas as pd
import seaborn as sb

# seaborn 패키지의 iris 데이터로 데이터 프레임 생성
df_iris = sb.load_dataset('iris')
df_iris
```

```python
df_iris['setosa_PL1.5'] = np.where( (df_iris['petal_length'] < 1.5 ) & ( df_iris['species'] =='setosa' ),
                         1, 0)

df_iris
```

- Q2) R의 iris 데이터에서 Species와 Petal.Length가 1.5 미만인지의 여부로 데이터를 그룹지어 네 개 변수 (Sepal.Length, Sepal.Width, Petal.Length, Petal.Width)의 평균을 출력하라.

```python
df_iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].mean()
```

```python
df_iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].groupby(df_iris['petal_length'] < 1.5).mean()
```

#### [Problem]  transform
- baseball 데이터는 1871년부터 2007년까지 총 1,228명의 미국 야구선수들의 타격에 대한 정보가 저장된 데이터이다. 각 선수는 id 칼럼으로 구분되어 있으며, 총 22개의 변수와 21,699개의 행으로 이루어져 있다.'g’ 칼럼은 각 선수가 해당 연도에 출전한 게임 수를 나타낸다.

```python
# baseball 데이터 불러오기 및 데이터 프레임 생성
import pandas as pd
df_baseball = pd.read_csv( "C:/Users/Administrator/GitHub/TIL/ADP_study/rawdata/baseball.csv")
df_baseball
```
