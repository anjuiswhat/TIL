## 4. 조건별 요약

```python
## Create Company Column using split
sample_data["Company"] = sample_data["brand"].str.split(' ').str[0]
```

```python
# Company별 hp 평균 (1개 사) 이 또한 .sum .max 등 사용 가능
sample_data["hp"][sample_data['Company']=='Merc'].mean()
# Company별 차종 수
sample_data["Company"].value_counts()
```

```python
## using Groupby ## 
sample_data.groupby(['Company']).groups.keys()
# Group별 길이
len(sample_data.groupby(['Company']).groups['Ford'])
# 모든 Com별 hp 평균 .sum .max .count 등 사용 가능
sample_data.groupby('Company')['hp'].mean()
# 모든 Com&am별 hp 평균
sample_data.groupby(['Company','am'])['hp'].mean()
sample_data.groupby(['Company','am'], as_index=False)['hp'].mean() #df화
```

## 5. apply 함수 적용

```python
#Col별 합계
sample_data.apply(np.sum, axis=0)
```

```python
# create numeric dataset
sample_num = sample_data[["hp", "cyl", "disp"]]
```

```python
# apply 함수 적용
sample_num.apply(np.square) #제곱
sample_num.apply(sum) #col별 합계
sample_num.apply(sum, axis = 1) #row별 합계

##특수 함수 생성 후 적용 (원값 * 3 + 1  구하기)
def col_sum(x,a,b):
    return a*x+b

col_sum(sample_num, 3, 1)


def size_test(x) :
    if x>150 :return "Big"
    else : return "Small"

sample_num_2 = sample_num.copy()    
sample_num_2["Size"] = sample_num_2['hp'].apply(size_test)    
```

## 6. crosstab

```python
pd.crosstab(sample_data.am, sample_data.vs) # contingency table
pd.crosstab(sample_data.Company, [sample_data.am, sample_data.vs]) 
pd.crosstab(sample_data.am, sample_data.vs, margins = True) #행열별 합계
pd.crosstab(sample_data.am, sample_data.vs, normalize = True) #전체를 1로 둔 비율
pd.crosstab(sample_data.am, sample_data.vs, normalize = 'index') #Row합 1
pd.crosstab(sample_data.am, sample_data.vs, normalize = 'columns') #Col합 1

```
