## SARIMAX(Seasonal ARIMAX with eXogenous model)

###### 계절성 변동의 제거등을 포함하여 시스템 식별을 수행하는 모델

###### trend에 대하여 ARIMA를 수행하고, 계절성에 대해서 추가적으로 ARIMA를 수행

출처 : https://sosoeasy.tistory.com/397

trend에 대한 부분(p,d,q)과 seasonal에 대한 부분(P,D,Q,m)


### 파라미터 선택방법

##### p : pacf에서 수렴하기 직전값. ( =2 )

##### d : adf테스트 및 추세의 가시적 확인으로 추세여부 확인( =1 )

##### q : acf에서 수렴하기 직전값 ( =3 )

##### P : pacf에서 계절성이 주기가 몇 번 반복 됐는지 확인 ( =3 )

##### D : 계절성이 있는지 확인하고 계절성 차분의 필요성에 따라 1 or 0 ( =1 )

##### Q : acf기준 계절성 주기가 몇 번 반복 됐는지 확인 ( =3 )

##### m : 계절성 주기 ( =24 )

To perform forecasting using the ARIMA model, we required a stationary time series.

If the data is not stationary, we can do one thing: either make the data stationary or use the SARIMAX model.

To know more about the time series stationarity, we can perform the ADfuller test, a test based on hypothesis, where if the p-value is less than 0.05, then we can consider the time series is stationary, and if the P-value is greater than 0.05, then the time series is non-stationary.

```python
from statsmodels.tsa.stattools import adfuller
dftest = adfuller(df.Sales, autolag = 'AIC')
print("1. ADF : ",dftest[0])
print("2. P-Value : ", dftest[1])
print("3. Num Of Lags : ", dftest[2])
print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
print("5. Critical Values :")
for key, val in dftest[4].items():
    print("\t",key, ": ", val)
```

Here we can see that the p-value is higher for our dataset, and we can say that the evidence of the null hypothesis is low; hence the time series is non-stationary

```python
rolling_mean = df.rolling(window = 12).mean()
df['rolling_mean_diff'] = rolling_mean - rolling_mean.shift()
ax1 = plt.subplot()
df['rolling_mean_diff'].plot(title='after rolling mean & differencing');
ax2 = plt.subplot()
df.plot(title='original');
```

```python
dftest = adfuller(df['rolling_mean_diff'].dropna(), autolag = 'AIC')
print("1. ADF : ",dftest[0])
print("2. P-Value : ", dftest[1])
print("3. Num Of Lags : ", dftest[2])
print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
print("5. Critical Values :")
for key, val in dftest[4].items():
  print("\t",key, ": ", val)
```

## 주가 데이터 시계열 분석

```python
url = 'https://sites.google.com/site/datasciencehiro/datasets/Stock_7203.txt'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
```

```python
df = pd.read_csv(url, parse_dates = [0])
```

```python
df.columns = ['날짜','시가','고가','저가','종가','거래량','매매대금']
```

```python
df.index = df['날짜']
```

```python
df = df[df.columns.difference(['날짜'])].sort_index()
```

```python
df.head()
```

### 이동평균

### 지수 이동평균
##### 최근 주가를 더 잘 반영하기 위해 최근 주가에 가중치를 더 주어 계산

```python
item = '종가'
ax = df[item].plot(color = 'black',label = item)
W = 25
rol = df[item].rolling(window=W).mean()
rol.plot(ax = ax , label = '이동평균25일', color = 'red', linestyle = 'dashed')
ewm = df[item].ewm(span=W).mean()
ewm.plot(ax = ax, label='지수이동평균스팬25', color= 'green', linestyle = 'dashed')
plt.grid()
plt.legend(loc='upper right')
```

W 값이 클수록 고주파 성분을 줄이고, 위상이 더 지연됨.

### 볼린저밴드

B(k) = k일의 이동평균 +- k일의 표준편차(sigma(k)) * m

2개의 시계열 데이터가 되고, 이 데이터 사이의 영역을 볼린저 밴드라고 한다.

```python
item = '종가'
W = 25
ax = df[item].plot(label = item, color = 'black', linestyle = 'solid')
rolm = df[item].rolling(window=W).mean()
rolst = df[item].rolling(window=W).std()
m = 2
upper_band = rolm + rolst * m
lower_band = rolm - rolst * m
upper_band.plot(ax = ax, label='Upper band', color= 'green', linestyle = 'dashed')
lower_band.plot(ax = ax, label='Lower band', color= 'red', linestyle = 'dashed')
plt.grid()
plt.legend(loc='upper right')
```
