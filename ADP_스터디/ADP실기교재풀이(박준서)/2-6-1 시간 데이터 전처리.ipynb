## 2. import time series data

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'])
sample_df = df
```

```python
# 숫자를 입력하여 시간 형태로 저장
d = datetime.datetime(2018, 5, 19)
# 데이터를 원하는 형태로 출력
d.strftime('%m-%d-%Y')
```

```python
Date = sample_df['date']
sample_df['date'] = Date.dt.strftime('%d-%b-%Y') # 연월일
Date = pd.to_datetime(Date)
sample_df['date'] = Date.dt.strftime('%Y-%m') # 연월일 중 일부만 사용시 pd.to_datetime 변경 후 다시 strftime
```

```python
sample_df_1 = sample_df.set_index('date')
sample_df_1['value'].plot(title = 'blah')
```
