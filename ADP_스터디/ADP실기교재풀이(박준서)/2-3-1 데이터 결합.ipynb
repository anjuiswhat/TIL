## 1. import dataframe 

### 1. import libraries

```python
import pandas as pd
import numpy as np
```

## 2. import mtcars data

```python
data = pd.read_csv('https://gist.githubusercontent.com/ZeccaLehn/4e06d2575eb9589dbe8c365d61cb056c/raw/64f1660f38ef523b2a1a13be77b002b98665cdfe/mtcars.csv')
data.rename(columns={'Unnamed: 0':'brand'}, inplace=True)
```

```python
sample_data = data
```

## 3. 데이터 세로로 결합 (merge rows)

```python
  ## 데이터 분할 ##
sub_1_data = sample_data.loc[sample_data.brand.str[0]=='F']
sub_2_data = sample_data.loc[sample_data.brand.str[0]=='M']
```

```python
### concatenate
pd.concat([sub_1_data, sub_2_data])
##np.concatenate([ar1, ar2], axis=0) for np data ##
```

## 4. 데이터 가로로 결합 (merge cols)

```python
  ## 데이터 분할 ##
sub_3_data = sample_data[['disp', 'hp', 'drat']]
sub_4_data = sample_data[['mpg', 'cyl', 'wt']]
```

```python
pd.concat([sub_3_data, sub_4_data], axis = 1)
```

## 5. 데이터 join

### inner join 

```python
   ###데이터 분할##
sub_5_data = sample_data.loc[sample_data.brand.str[0]=='F']
sub_6_data = sample_data.loc[sample_data.brand.str[0]=='V']
sub_7_data = pd.concat([sub_5_data, sub_6_data], axis = 0)
sub_5_data_2 = sub_5_data[['wt', 'qsec','brand', 'vs']]
sub_7_data_2 = sub_7_data[['cyl','brand', 'disp', 'drat']]
```

```python
# inner join by key column #
pd.merge(sub_5_data_2,sub_7_data_2, left_on='brand', right_on='brand', how='inner')
# inner join by index #
#pd.merge(sub_5_data_2,sub_7_data_2, left_index=True, right_index=True, how='inner')
```

### outer join 

```python
sub_8_data = sample_data.loc[sample_data.brand.str[0]=='A']
sub_9_data = sample_data.loc[sample_data.brand.str[0]=='B']
sub_10_data = sample_data.loc[sample_data.brand.str[0]=='C']
sub_11_data = pd.concat([sub_8_data, sub_9_data], axis = 0)
sub_12_data = pd.concat([sub_10_data, sub_9_data], axis = 0)
sub_11_data_2 = sub_11_data[['wt', 'qsec','brand', 'vs']]
sub_12_data_2 = sub_12_data[['cyl','brand', 'disp', 'drat']]
```

```python
pd.merge(sub_11_data_2, sub_12_data_2, left_on='brand', right_on='brand', how='outer')
```
