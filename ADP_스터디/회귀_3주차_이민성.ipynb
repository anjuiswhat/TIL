# 이상치 제거

```python
# 이상치를 제거한 데이터프레임을 생성하는 함수 

def get_outlier_drop( df=None, outlier_columns=None, weight=1.5):

    outlier_index = list()
    df_outlier_drop = df.copy()
    
    
    for i in range(0, outlier_columns.shape[1]):  # 분류분석의 경우 'Class' 변수를 제외하고 이상치를 탐색하기 위해 범위를 조정할 필요가 있다.
                                     # 종속변수 'Class'의 경우, 분포 불균형으로 인해 1을 모두 이상치로 인식함.  
        
        # col에 해당하는 column 데이터만 추출, 1/4 분위와 3/4 분위 지점을 np.percentile로 구함. 
        col = outlier_columns.iloc[:,i]
        quantile_25 = col.quantile(0.25)
        quantile_75 = col.quantile(0.75)
        
        # IQR을 구하고, IQR에 1.5를 곱하여 최대값과 최소값 지점 구함. 
        iqr = quantile_75 - quantile_25
        iqr_weight = iqr * weight
        lowest_val = quantile_25 - iqr_weight
        highest_val = quantile_75 + iqr_weight
        
        # 최대값 보다 크거나, 최소값 보다 작은 값을 이상치로 설정하고 이상치와 해당 index를 반환.
        outlier = col[ (col < lowest_val) | (col > highest_val) ]
        
        # 이상치가 존재하는 행 index 주소를 리스트에 추가
        if len(outlier) == 0:
            pass
        else:
            outlier_index.extend( outlier.index )
            
    # 이상치가 존재하는 행을 모두 제거한 데이터프레임 생성
    df_outlier_drop = df_outlier_drop.drop( np.unique(outlier_index), axis=0 )
        
    return df_outlier_drop
```

# 회귀분석 수행

- y = α + β1·x1 + β2·x2 + ... + βp·xp + ε


- 귀무가설 H0 : β1 = β2 = ... = βp = 0
- 대립가설 H1 : β1 ~ βp 중 적어도 하나 이상의 계수는 0 이 아니다.

## [ 참고 ] `statsmodels`을 사용한 회귀분석 시 유의사항

- statsmodels 패키지는 sklearn 과는 달리 회귀분석 결과와 관련된 수치들을 손쉽게 출력할 수 있는 방법을 제공하지만 해당 방법들의 대다수가 회귀모델 적합 시 생성된 값들, 즉, training data로 학습/예측한 결과만 확인할 수 있다는 단점이 있다.
  
- 예를 들어 적합된 모델로 training data에 대한 '예측값', '잔차', 'MSE', 'R-squared', 'Adj. R-squared' 등을 바로 출력하는 메서드나 속성을 지원하지만, test data에 대한 해당 통계량에 대해서는 지원하지 않는다.

- 즉, statsmodels의 많은 메서드/속성들이 'training data'를 사용하여 적합된 회귀모델에 'test data'를 적용하여 예측한 결과에 대한 출력은 지원하지 않으며, 이 경우에는 결국 sklearn 패키지로 만든 회귀분석 결과를 검토했을 때와 마찬가지로 별도의 코드를 작성하여 적용해야 한다.

    - Returns : `RegressionResults`
      - https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.RegressionResults.html
      - Attributes:
        - 예측값(Training Data) : `.fittedvalues`
        - 잔차 : `.resid`

        - 회귀모델의 F-통계량 및 p-value : `.fvalue`, `.f_pvalue`
        - 회귀계수 : `.params`
        - 회귀계수의 Standard Error : `.bse`

        - `.predict( exog=None, transform=True, *args, **kwargs )` : 적합된 모델을 사용하여 반응변수 예측
          - https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.RegressionResults.predict.html
        - `.summary( yname=None, xname=None, title=None, alpha=0.05 )` : 회귀 결과 요약
          - https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.RegressionResults.summary.html
        - `.get_prediction( exog=None, transform=True, weights=None, row_labels=None, **kwargs )`
          - https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.RegressionResults.get_prediction.html

### 평가 지표

- MSE (Mean Squared Error) : 실제값과 예측값의 차이를 제곱하여 평균 (오류의 제곱을 구하므로 실제 오류 평균보다 더 커지는 특성이 있음)


- RMSE (Root Mean Squared Error) : MSE에 루트를 씌운 값 
  - `sklearn.metrics.mean_squared_error( y_true, y_pred, *, sample_weight=None, multioutput='uniform_average', squared=True )`  
    - https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html
    - MSE : squared = True (기본값)
    - RMSE : squared = False


- MAE (Mean Absolute Error) : 실제값과 예측값의 차이를 절대값으로 변환하여 평균
  - `sklearn.metrics.mean_absolute_error( y_true, y_pred, *, sample_weight=None, multioutput='uniform_average' )`
    - https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_absolute_error.html


```python
# R-squared
r2_train = fit_model.rsquared  # r2_score( y_train, pred_train ) 또는 model.score( x_train, y_train )도 가능
r2_test = r2_score( y_test, pred_test )  # 또는 model.score( x_test, y_test )

# MSE (Mean Squared Error)
mse_train = fit_model.mse_resid  # mean_squared_error( y_train, pred_train )도 가능. 약간의 수치 차이는 있음.
mse_test = mean_squared_error( y_test, pred_test )

# RMSE (Root Mean Squared Error)
rmse_train = np.sqrt( fit_model.mse_resid )  # mean_squared_error( y_train, pred_train, squared=False )도 가능. 약간의 수치 차이는 있음.
rmse_test = mean_squared_error( y_test, pred_test, squared=False )

# MAE (Mean Absolute Error)
mae_train = mean_absolute_error( y_train, pred_train )
mae_test = mean_absolute_error( y_test, pred_test )

print( '< 평가지표 (Training Data) >')
print( 'R-squared: %.4f' % r2_train )
print( 'RMSE: %.4f' % rmse_train )
print( 'MSE: %.4f' % mse_train )
print( 'MAE: %.4f' % mae_train )
print()
print( '< 평가지표 (Test Data) >')
print( 'R-squared: %.4f' % r2_test )
print( 'RMSE: %.4f' % rmse_test )
print( 'MSE: %.4f' % mse_test )
print( 'MAE: %.4f' % mae_test )
print()
```

### Condition Number (상태지수; 조건수)

- 조건수(condition number)는 argument에서의 작은 변화의 비율에 대해 함수가 얼마나 변화할 수 있는지 에 대한 argument measure이다.
- 조건수가 크면 약간의 오차만 있어도 해가 전혀 다른 값을 가진다. 따라서 조건수가 크면 회귀분석을 사용한 예측값도 오차가 커지게 된다.


- 회귀분석에서 Condition Number가 커지는 원인은 크게 두 가지가 있다.  
  - 변수들의 단위 차이로 인해 숫자의 스케일이 크게 달라지는 경우. 이 경우에는 스케일링(scaling)으로 해결한다.  
  - 다중 공선성 즉, 상관관계가 큰 독립 변수들이 있는 경우. 이 경우에는 변수 선택이나 PCA를 사용한 차원 축소 등으로 해결한다.


- 다중공선성 측면만 본다면 (스케일링 측면의 문제가 없다고 가정하면)  
  - 상태지수가 10보다 크면 다중공선성이 어느 정도 존재 하는 것으로 간주.  
  - 상태지수가 30보다 크면 다중공선성이 매우 심각한 것으로 간주.

### 오차의 독립성 (Durbin-Watson 통계량)

- 오차들은 서로 독립적이라는 가정.

  - 2 에 가까우면 오차항의 독립성 만족,

  - 4 에 가까우면 음(-)의 자기상관관계
  - 자기상관성은 흔히 시간에 따라 관측된 시계열 데이터에서 많이 발생하는데, 자기상관성이 있는 것으로 판단되면, ARMA(자기회귀 이동평균)와 같은 시계열데이터 분석방법을 이용하는 것이 바람직하다.

  - 독립변수의 갯수가 많을 때, 특정한 하나의 독립변수의 영향력을 시각화하는 방법
  - Added Variable Plot이라고도 한다.
  - 부분회귀 플롯을 그리기 위해서는 3번의 선형 회귀분석을 해야 한다.
    - 1) 특정한 독립변수 x2를 제외한 나머지 독립변수 X1들로 종속변수 y를 선형 회귀분석하여 잔차 y'를 구한다.
    - 2) 특정한 독립변수 x2를 제외한 나머지 독립변수 X1들로 특정한 독립변수 x2를 선형 회귀분석하여 잔차 x2'를 구한다.
    - 3) 잔차 x2'를 독립변수로, 잔차 y'를 종속변수로 하여 선형 회귀분석한다.
    - 이렇게 구한 x2', y'의 스캐터 플롯과 회귀분석 결과를 나타낸 것이 부분회귀 플롯이다.

  - CCPR 플롯은 이 성분 중에서 xi 를 가로축으로 (wixi + e)를 세로축으로 그린 스캐터 플롯이다.
