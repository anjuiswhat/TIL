### RMSLE

- target 값을 Log 변환하여 회귀분석을 수행했을 경우에 RMSE를 구하면 오류값의 로그에 대한 RMSE 가 구해짐

### RMSE

- Log 변환된 target 기반으로 학습/예측된 값에 expm1 을 적용하여 원래대로 스케일 변환하여 RMSE 산출

```python
y_pred = fit_model.predict(x_test)

y_pred = np.expm1(y_pred)
y_test = np.expm1(y_test)

rmse = mean_squared_error(y_test, y_pred, squared = False)

print('RMSE =', rmse)
print()
```
