#### wordcloud 라이브러리 설치

```python
from konlpy.tag import Okt, Kkma, Komoran #형태소 분석
from collections import Counter #빈도수 세기
from wordcloud import WordCloud
```

java.nio.file.InvalidPathException 에러가 나는 경우 jpype1을 0점대 버전으로 재설치 pip install "jpype1<1"

```python
ok = Okt()
kkma = Kkma()
kom = Komoran()
```

```python
result = []
for i in text:
    for s in ok.nouns(str(i)):
        result.append(s)
```

```python
res_count = Counter(result)
```

```python
word_data = res_count.most_common(100) #상위 100개 단어 추출
```

```python
word_data[:5]
```

```python
wc = WordCloud()
cloud = wc.generate_from_frequencies(dict(word_data))
plt.figure(figsize = (10, 10))
plt.imshow(cloud)
plt.show()
```
