### 1. import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## 2. import mtcars data

```python
data = pd.read_csv('https://gist.githubusercontent.com/ZeccaLehn/4e06d2575eb9589dbe8c365d61cb056c/raw/64f1660f38ef523b2a1a13be77b002b98665cdfe/mtcars.csv')
data.rename(columns={'Unnamed: 0':'brand'}, inplace=True)
```

```python
sample_data = data
```

## 3. boxplot / 4분위수

```python
## 개별 Column boxplot##
sns.boxplot(y=sample_data["mpg"])
## 그룹별 boxplot ##
sns.boxplot(y=sample_data["mpg"], x=sample_data["vs"])
# 그룹 + x boxplot ##
sns.boxplot(y="mpg", x= "vs", hue = "cyl", data = sample_data)
```

```python
# outlier 확인
sample_data['disp'].quantile(0.25) # 0.5 0.75 로 다른 quant 확인
IQR = sample_data['disp'].quantile(0.75) - sample_data['disp'].quantile(0.25)
outlier = sample_data['disp'].quantile(0.75) + 1.5*IQR # 이상치 기준값
sample_data[sample_data['disp'] > outlier]

# outlier인 행 삭제
outliers = sample_data['disp'].quantile(0.75) + 0.1*IQR # 이상치 기준값
sample_data_1 = sample_data.drop(sample_data[sample_data["disp"]>outliers].index)
sample_data_1
```
