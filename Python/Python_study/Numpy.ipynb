###  Indexing & Slicing
#### - ndarray       .  List    .
#### -    (view)  .
####        copy() Method    (fancy indexing)  .

###    (fancy indexing)
#### -      (boolean indexing)   .

```python
a = np.array([1, 2, 3, 4, 5])
b = a[0: 2]
print(a, b)
print()

b[0] = 10
print(a, b)
print()

a = np.array([1, 2, 3, 4, 5])
b = a[0:2].copy()
print(a, b)
print()

b[0] = 10
print(a, b)
```

###    (integer indexing)
#### -    a[[row1, row2], [col1, co12]]    .

```python
a = np.array([1, 2, 3, 4, 5])
print(a[0], a[:2], a[::2], a[3::-1])
print()

b = a[[0, 2]]
b[0] = 10
print(a, b)
```

###    (boolean indexing)
#### -     .

```python
a = np.array([1, 2, 3, 4, 5])
b = a[[True, False, True, False, False]]
c = a[a > 2]
print(b)
print()
print(a > 2)
print()
print(c)
```

###   Method

```python
a = np.array([3, 2, 1, 4, 2, 1])
b = np.array([8, 3, 5, 9, 4, 7])

print(np.unique(a))          #    (sorted)  
print()
print(np.intersect1d(a, b))  #  (sorted) 
print()
print(np.union1d(a, b))      #  (sorted) 
print()
print(np.in1d(a, b))         #  b   a    
print()
print(np.setdiff1d(a, b))    #  (sorted) 
print()
print(np.setxor1d(a, b))     # (  ) (sorted) 
```
