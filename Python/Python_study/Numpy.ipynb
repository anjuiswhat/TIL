### ◇ Indexing & Slicing
#### - ndarray의 요소나 일부분을 선택할 때 대상을 지정하는 방법이다. 파이썬 List 의 인덱싱과 기능이 비슷하다.
#### - 기본적으로 요소의 참조인 뷰(view) 를 반환한다.
####   만약 값의 반환이 필요한 경우 copy() Method 나 팬시 인덱싱 (fancy indexing) 을 이용한다.

### ◇ 팬시 인덱싱 (fancy indexing)
#### - 요소 선택 범위룰 배열이나 논리값 (boolean indexing) 으로 지정하는 방법이다.

```python
a = np.array([1, 2, 3, 4, 5])
b = a[0: 2]
print(a, b)
print()

b[0] = 10
print(a, b)
print()

a = np.array([1, 2, 3, 4, 5])
b = a[0:2].copy()
print(a, b)
print()

b[0] = 10
print(a, b)
```

### ◇ 정수 인덱싱 (integer indexing)
#### - 요소 선택 범위를 a[[row1, row2], [col1, co12]] 와 같이 지정하는 방법이다.

```python
a = np.array([1, 2, 3, 4, 5])
print(a[0], a[:2], a[::2], a[3::-1])
print()

b = a[[0, 2]]
b[0] = 10
print(a, b)
```

### ◇ 부울린 인덱싱 (boolean indexing)
#### - 요소의 선택을 논리값으로 지정하는 방법이다.

```python
a = np.array([1, 2, 3, 4, 5])
b = a[[True, False, True, False, False]]
c = a[a > 2]
print(b)
print()
print(a > 2)
print()
print(c)
```

### ◇ 연산
#### - 행렬의 요소별 (element-wise) 연산인 벡터화 연산(vectorized operation)을 지원하여 속도가 빠르다.
#### - 배열의 원소에 적용되는 연산으로 +, -, *, / 및 add, subtract, multiply, divide 등의 함수를 사용할 수 있다.
#### - Shape 가 같은 두 배열은 요소별로 연산이 수행된다 .
####   만약 두 배열 Shape 가 다른 경우, 
####   대상 배열 각각의 행과 열의 최대 크기로 Shape 를 일치화 시키는
####   브로드캐스팅(Broadcasting) 이 일어난다.

```python
a = np.array(range(1, 7)). reshape(2, 3)
b = np.ones((2, 3))

d1 = a+1
d2 = a-1
d3 = a*5
d4 = a/2
print(d1)
print()
print(d2)
print()
print(d3)
print()
print(d4)
```

### ◇ 집합 Method

```python
a = np.array([3, 2, 1, 4, 2, 1])
b = np.array([8, 3, 5, 9, 4, 7])

print(np.unique(a))          # 중복올 제거한 후 정렬(sorted) 하여 반환
print()
print(np.intersect1d(a, b))  # 교집합을 정렬(sorted)하여 반환
print()
print(np.union1d(a, b))      # 합집합을 정렬(sorted)하여 반환
print()
print(np.in1d(a, b))         # 집합 b에 포함되는 집합 a의 원소를 논리 값으로 반환
print()
print(np.setdiff1d(a, b))    # 차집합 정렬(sorted)하여 반환
print()
print(np.setxor1d(a, b))     # 대칭차집합(합집합에서 교집합을 제외) 정렬(sorted)하여 반환
```

### ◇ 기타 Method
#### - 원소별로 연산을 하는 유니버셜 함수(unfunc) 와 필터링하는 제공한다. 대표적인 기능은 다음과 같다.
.


```python
a= np.array([-1.7, -1.5, -0.3, 0, 0.3, 1.5, 1.7])
np.ceil(a)
np.floor(a)
np.rint(a)

a = np.arange(20)

condition = a < 3
np.extract(condition, a)
condlist = [a< 7, a> 15]
choicelist = [a, a*2]

print(condlist)
print()
print(choicelist)
print()
print(np.select(condlist, choicelist))
print()
print(np.where(a % 2 == 0, '짝수', '홀수'))
```
