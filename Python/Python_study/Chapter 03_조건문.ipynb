# 03-1. 불(Bool) 자료형과 if 조건문
불(Bool),  비교 연산자,  논리 연산자,  if 조건문

```python
# 불(bool) 만들기 : 비교 연산자
# 불(bool) 자료형이란 참(True)과 거짓(False)을 나타내는 자료형이다.
# 불 자료형은 True, False 2가지 값만을 가질 수 있다.
# == 같다, != 다르다, <= 작거나 같다, >= 크거나 같다 
# 문자열에도 비교 연산자를 적용할 수 있다. (가나다 또는 abc 순서에서 앞에 있는 문자가 작은 값을 가짐.)

print("10 == 100 → ",10 == 100)
print("10 != 100 → ",10 != 100)
print("10 < 100  → ",10 < 100)
print("10 <= 100 → ",10 <= 100)
print("10 > 100  → ",10 > 100)
print("10 >= 100 → ",10 >= 100)
print()

print('"가방" =< "가방" → ',"가방" == "가방")
print('"가방" != "하마" → ',"가방" != "하마")
print('"가방" < "하마" → ',"가방" < "하마")
print('"가방" > "하마" → ',"가방" > "하마")
print('"a" > "b" → ',"a" > "b")
print()

x=25
print('''x=25
10 < x < 30 → ''',10 < x < 30)
```

```python
# 불(bool) 연산하기 : 논리 연산자
# not : 불을 반대로 전환

print("< not 연산자 >")
print("not True  → ", not True)
print("not False → ", not False)
print()

x = 10
under_20 = x < 20
print("under_20:", under_20)
print("not under_20:", not under_20)
print("-------------------------")
```

```python
# and : 피연산자 두 개가 모두 참일 때 True를 출력, 그 외에는 모두 False를 출력
# or  : 피연산자 두 개 중에 하나만 참이라도 True를 출력, 두 개가 모두 거짓일 때만 False를 출력

print("< and 연산자 >")
print("True  'and' True  → ", True and True)
print("True  'and' False → ", True and False)
print("False 'and' True  → ", False and True)
print("False 'and' False → ", False and False)
print()
print("< or 연산자 >")
print("True  'or' True  → ", True or True)
print("True  'or' False → ", True or False)
print("False 'or' True  → ", False or True)
print("False 'or' False → ", False or False)
```

```python
# if 조건문 : 조건에 따라 코드의 실행 흐름 변경
# if 다음에 오는 조건이 참인 경우만 실행됨.
# if 조건문 뒤에는 반드시 콜론(:)을 붙여야 함.

# if 뒤 조건이 True 이므로 다음 코드를 실행함.
if True:
    print("True 입니다.") 

# if 뒤 조건이 Flase 이므로 다음 코드를 실행하지 않음.
if False:
    print("False 입니다.")
```

```python
# if문을 이용하여 입력한 정수의 양수, 음수 판단

# 입력을 받습니다.
number = input("정수 입력> ")

# 문자열로 입력 받은 숫자를 정수 타입으로 변환
number = int(number)

# 양수 조건
if number > 0:
    print("양수입니다")

# 음수 조건
if number < 0:
    print("음수입니다")

# 0 조건
if number == 0:
    print("0입니다")
```

```python
# 짝수와 홀수 구분하는 코드 (1)

# 입력을 받습니다.
n = input("정수 입력> ")

# 입력한 숫자가 정수인지 판별
if not n.isdecimal():
    print("입력된 숫자는 정수가 아닙니다!!!")
if n.isdecimal():
    
    # 입력받은 숫자의 마지막 자리 문자 구하기
    last_character = n[-1]
    
    # 입력받은 자료는 문자열임. 마지막 자리 문자를 숫자형(정수)로 변환    
    last_number = int(last_character)  
    
    if last_number == 0 \
    or last_number == 2 \
    or last_number == 4 \
    or last_number == 6 \
    or last_number == 8 :
        print("짝수 입니다.")
        
    if last_number == 1 \
    or last_number == 3 \
    or last_number == 5 \
    or last_number == 7 \
    or last_number == 9:
        print("홀수 입니다.")
```

```python
# 짝수와 홀수 구분하는 코드 (2) : in 연산자 사용

# 입력을 받습니다.
n = input("정수 입력> ")

# 입력한 숫자가 정수인지 판별
if not n.isdecimal():
    print("입력된 숫자는 정수가 아닙니다!!!")
if n.isdecimal():
    
    # 입력받은 숫자의 마지막 자리 문자 구하기
    last_character = n[-1]
    
    if last_character in "02468":
        print("짝수 입니다.")
    
    if last_character in "135749":
        print("홀수 입니다.")
```

```python
# 짝수와 홀수 구분하는 코드 (3) : 나머지 연산자 % 사용

# 입력을 받습니다.
n = input("정수 입력> ")

# 입력한 숫자가 정수인지 판별
if not n.isdecimal():
    print("입력된 숫자는 정수가 아닙니다!!!")
if n.isdecimal():
    
    #입력한 숫자를 정수로 변환한 후, 2 로 나눈 나머지에 따라 문구 출력
    number = int(n)
    if number % 2 == 0:
        print("짝수 입니다.")
    if number % 2 != 0:
        print("홀수 입니다.") 
```

```python
# 짝수와 홀수 구분하는 코드 (4) : if ~ else 구문 사용

# if 조건:
#     조건이 참일때 실행할 코드
# else:
#     조건이 거짓일때 실행할 코드

# 입력을 받습니다.
n = input("정수 입력> ")

# 입력한 숫자가 정수인지 판별
if not n.isdecimal():
    print("입력된 숫자는 정수가 아닙니다!!!")
else:
    #입력한 숫자를 정수로 변환한 후, 2 로 나눈 나머지에 따라 문구 출력
    number = int(n)
    if number % 2 == 0:
        print("짝수 입니다.")
    else:
        print("홀수 입니다.") 
```
