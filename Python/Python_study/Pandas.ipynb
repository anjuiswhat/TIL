### ◇ Indexing & Slicing
#### - 팬시 인덱싱 등의 특징과 기능이 List 나 ndarray 와 동일하다.

```python
si = pd.Series([1, 2, 3, 4, 5],
               index = ['a', 'b', 'c', 'd', 'e'])

print(si[0], si['b'], si.get('g', 0))
print(si[: 3], si[2:3], si[-1], si[::2], si[::-1])


si1 = si[:3]
si1[0] = 11
print(si)

si2 = si[[True, False, True, False, True]]
print(si2)

si3 = si [si > 3]
print(si3)

si4 = si[(si >= 2) & (si <=4)]
print(si4)

print(si [['c', 'b']])
print(si ['c':'e'])
print(si ['c':'e':2])
print(si['e':'b':-1]) 
```

### ◇ 연산
#### - 벡터화 연산(vectorized operation)을 지원하고 ndarray 와 기능이 동일하다.

#### 연산자

```python
sp = pd.Series([1, 2, 3, 4, 5])
sp1 = pd.Series([1, 2, 3, 4, 5])

d1 = sp + 1
d2 = sp - 1
d3 = sp * 5
d4 = sp / 2
d5 = sp + sp1
d6 = sp - sp1
d7 = sp * sp1
d8 = sp / sp1
d9 = sp % sp1
print(d1, d2, d3, d4, d5, d6, d7, d8, d9)
```

#### Method

```python
d10 = pd.Series.add(sp, sp1)
d11 = pd.Series.sub(sp, sp1)
d12 = pd.Series.mul(sp, sp1)
d13 = pd.Series.div(sp, sp1)
```
